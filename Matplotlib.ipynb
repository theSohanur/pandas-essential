```python
x = [1,2,3,4]
y = [11,22,33,44]
plt.plot(x,y)
```

```python
# 1st method
fig = plt.figure() #creates a figure
ax = fig.add_subplot() # adds some axis
plt.show()
```

```python
# 2nd method 
fig = plt.figure() #creates a figure
ax = fig.add_axes([1,1,1,1]) 
ax.plot(x,y)# adds some data
plt.show()
```

```python
# 3rd method
fig, ax = plt.subplots()
ax.plot(x,[50,100,200,250]);
```

## Making figures with NumPy arrays

We want : 
* Line Plot
* Scatter plot
* Bar Plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
#Create some data 
x = np.linspace(0,10,100)
x
```

```python
x[:10]
```

```python
#Plot the data and create line plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Use same data make a scatter 
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Make a plot from Dictionary
nut_butter_prices = {
    "Almond butter":10,
    "Peanut butter":8,
    "Cashew butter":12
}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values()); #(x,y)
ax.set(title = "Nut Butter Store",
      ylabel = "Price($)");
```
